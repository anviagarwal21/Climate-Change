```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing data
os.getcwd()
os.chdir("/Users/anviagarwal/downloads")
climate=pd.read_csv('climate_change.csv')
```

```python
#Change the 'date' column from object to date time
climate['date']=pd.to_datetime(climate['date'])
#'date' column is now in the YYYY-MM-DD format
climate.info()
```

```python
#Time graph: Plotting CO2 emmisions over time
fig, ax= plt.subplots()
ax.plot(climate['date'], climate['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 Emissions')
ax.set_title('CO2 Emisions Over Time')
plt.show()
#Seasonality is seen - in a year the emmissions changed based on the seasons
#No cyclical pattern
#Clear upwards trend
```

```python
#Time graph: relative temperature over time
fig, ax= plt.subplots()
ax.plot(climate['date'], climate['relative_temp'], color='orange')
ax.set_xlabel('Time')
ax.set_ylabel('Relative Temperature')
ax.set_title('Relative Temperature Over Time')
plt.show()
#Much more volatile compared to co2
```

```python
#Zoom in on the 1990s CO2 emissions over time
#Syntax: df[(df['coln name']>'2000-01-01')&(df['coln name']<'1989-12-31')]
climate[(climate['date']<'2000-01-01')&(climate['date']>'1989-12-31')].head()
#Making it a different dataset called climate_90s
climate_90s=climate[(climate['date']<'2000-01-01')&(climate['date']>'1989-12-31')]

#Time graph of CO2 over time (only 1990s)
fig, ax= plt.subplots()
ax.plot(climate_90s['date'], climate_90s['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 Emissions')
ax.set_title('CO2 Emissions in 1990s')
plt.show()
```

```python
#Twin Axes
fig, ax= plt.subplots()
ax.plot(climate['date'], climate['co2'], color='red')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 Emissions', color='red')
ax.set_title('CO2 Emissions and Relative Temperature Over Time')
#Setting up twin axes
ax2 = ax.twinx()
ax2.plot(climate['date'], climate['relative_temp'], color = 'green')
ax2.set_ylabel('Relative Temperature', color='green')
```
